```python
import requests
from bs4 import BeautifulSoup
import csv
import re

BASE_URL = "http://www.paxnet.co.kr/tbbs/list"
stock_code = '086520'

# 페이지 범위 설정
start_page = 1
end_page = 1

# 헤더 설정
headers = {"User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36"}

with open('pax_all.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['제목', '조회수', '추천수', '작성일', '본문']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    # 헤더 쓰기
    writer.writeheader()

    for num_of_page in range(start_page, end_page + 1):

        # 파라미터 설정
        params = {'tbbsType' : 'L', 'id' : stock_code, 'page' : num_of_page}

        resp = requests.get(BASE_URL, params=params, headers=headers)

        soup = BeautifulSoup(resp.content, 'html.parser')
        contents = soup.find('ul', class_ = 'comm-list Basiclist').find_all('li')
        del contents[0]

        title_text = ''
        view = ''
        like = ''
        date_value = ''
        content_text = ''

        for i in contents:
            print('-'*15)
            # 제목 추출
            title_tag = i.find('div', class_='title')
            if title_tag:
                title = title_tag.find('p', class_='tit')
                if title:
                    title_text = title.text[:-4]
                    print("제목: " + title_text)
                else:
                    print("제목을 찾을 수 없습니다.")

            # 조회 수 추출
            view_tag = i.find('div', class_='viewer')
            if view_tag is None:
                pass
            else:
                view = view_tag.text.strip('조회 ')
                print("조회: " + view)
                
            # 추천 수 추출
            like_tag = i.find('div', class_='like')
            if like_tag is None:
                pass
            else:
                like = like_tag.text.strip('추천 ')
                print("추천: " + like)

            # 날짜 추출 
            date_tag = i.find('div', class_='date')
            if date_tag:
                date = date_tag.find('span', class_ = 'data-date-format')
                if date:
                    date_dict = date.attrs
                    date_value = date_dict.get('data-date-format')
                    print("작성일: " + date_value)
                else:
                    print("작성일: 작성일")

                    
            # 본문 추출
            if title_tag:
                link_tag = title_tag.find('p', class_ = 'tit').find('a', class_ = 'best-title')['href']
                link_id = re.sub(r'[^0-9]', "", link_tag)
#                 print(link_id)
                
                params_2 = {'id' : stock_code, 'seq' : link_id}
                resp_2 = requests.get('https://www.paxnet.co.kr/tbbs/view', params = params_2, headers = headers)
                soup_2 = BeautifulSoup(resp_2.content, 'html.parser')
                
                content_element = soup_2.select_one('div.board-view-cont')
                content_text = content_element.get_text()[:-73]
                print(content_text)
                print('-'*30)

            else:
                pass

            # 데이터를 CSV 파일에 추가
            writer.writerow({'제목': title_text, '조회수': view, '추천수': like, '작성일': date_value, '본문': content_text})

        # 현재 페이지 번호와 완료 메시지 출력
        print('-'*30)
        print(f"현재 페이지: {num_of_page} 페이지 크롤링 완료")    
```
